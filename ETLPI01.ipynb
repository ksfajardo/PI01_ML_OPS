# ETL with requested transformations

Importing the libraries I will use.

```python
import pandas as pd
import numpy as np
```

Transforming the cvs file into a DataFrame to get a general idea of what it looks like. 

```python
df=pd.read_csv('Datasets\movies_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

I am going to drop the columns that I will not be using (video, imdb_id, adult, original_title, vote_count, poster_path y homepage)

```python
df.drop(columns=['video','imdb_id','adult','original_title','vote_count','poster_path','homepage'], inplace=True)
```

Checking if there is any null values

```python
df.isna().sum()
```

I am going to replace the null values from the 'revenue' column with zeros. On the other hand, the column 'budget' has no null values, therefore I am going to leave it as is. 

```python
df['revenue']=df['revenue'].fillna(0)
```

Let's review the dates in the 'release date' column. We have 87 null values in this column, I am dropping those.  

```python
df=df.dropna(subset=['release_date'])
```

Now, I'm transforming the type of data on the 'release date' column to datetime. 

```python
df['release_date']=pd.to_datetime(df['release_date'],format='%Y-%m-%d',errors='coerce')
```

Checking to see if i still have null values after converting to datetime

```python
df['release_date'].isna().sum()
```

```python
df[df['release_date'].isna()]
```

Dropping the null values

```python
df=df.dropna(subset=['release_date'])
```

```python
df.release_date.isna().sum()
```

Creating a new column 'realase_year', getting the year from the 'release_date' column.

```python
df['release_year']=df['release_date'].dt.year
```

Now, I am goint to create a new column, 'return' and I need to divide the values in the revenue column by the values in the budget columns. To do this, I need to check that I can do mathematical operations with both columns.

```python
df.info()
```

Since the column budget is being treated as a string, I am going to change its data type to float to match the data type of the revenue column. 

```python
df['budget']=pd.to_numeric(df['budget'],downcast='float',errors='coerce')
```

```python
df['return']=df['revenue'].divide(df['budget']).fillna(0)
```

Some values were 'inf' after i did the operation, so I am going to replace those values with NaN and then fill those NaN with 0.

```python
df['return'].replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df['return'].fillna(0,inplace=True)
```

So, now that I have partially cleaned the DataFrame, I am going to take the data that is presented in some columns as ditionaries and lists of dictiories and add it as new columns.

Let's start with the 'belongs_to_collection' column. I am going to fill the NaN values with curly brackets so that I can use the map fuction with the eval function to get to python recognize the data as dictionaries, then I will use the json_normalize to get a DataFrame with the data in individual columns. 

PD: I did reset the index here since I had previosly dropped some rows, this is purely to keep the index in an ascending order.

```python
df['belongs_to_collection']=df['belongs_to_collection'].fillna('{}')
```

```python
df = df.reset_index(drop=True)
```

```python
df1=pd.json_normalize(df['belongs_to_collection'].map(eval))
```

```python
df1.head()
```

Now, I am going to drop the column "belongs_to_collection" and I am going to create two columns, one with the name of the collections and another with their id. I am leaving out the columns poster_path and backdrop_path, since they have virtually the same kind of data that I know I will not need. 

```python
df['collection_name']=df1['name']
df['idcollection']=df1['id']
df.drop(columns=['belongs_to_collection'],inplace=True)
df.head()
```

I am going to delete the additional df1 I created since I will not be needing it anymore.

```python
del df1
```

I am going to transform the column 'genres' into two columns, one with the genres id and another with the genres names.

Now, I am creating two functions to extract the names and id as lists. As you see, I did not use json_normalize this time because i figured it was easier and shorter this way.

```python
get_names = lambda lst: [d['name'] for d in lst]
get_id = lambda lst: [d['id'] for d in lst]
```

Now, I am going to use map in the values of column 'genres' and proceed to create new columns, filling them with the resulting list from applying the functions to the 'genres' column. 

```python
df['genres']=df['genres'].map(eval)
df['genres_name'] = df['genres'].apply(get_names)
df['idgenres'] = df['genres'].apply(get_id)
```

I drop the column 'genres'.

```python
df.drop(columns='genres',inplace=True)
df.head()
```

Next, I need to transform the data in the column 'spoken_languages'. For this, I am going to follow the same steps I did for the column 'genres' but I am going to create another the function so that it gets the iso_639_1 key from these lists of dictionaries.

```python
get_slcodes = lambda lst: [d['iso_639_1'] for d in lst]
```

```python
df['spoken_languages']=df['spoken_languages'].map(eval)
```

```python
df['slang_name'] = df['spoken_languages'].apply(get_names)
df['slang_isocode'] = df['spoken_languages'].apply(get_slcodes)
```

Now, I am going to drop the 'spoken_languages' column. 

```python
df.drop(columns='spoken_languages',inplace=True)
df.head()
```

Next, for the column 'production_companies' I am also reusing the previous functions get_names and get_id and following the same steps as before.

```python
df['production_companies']=df['production_companies'].map(eval)
```

```python
df['pcompany_name'] = df['production_companies'].apply(get_names)
df['idpcompany'] = df['production_companies'].apply(get_id)
```

```python
df.drop(columns='production_companies',inplace=True)
df.head()
```

Now, I am going to create a new function that gets the 'iso_3166_1' key from the last nested column,'production_countries' and follow the same spteps as before.

```python
get_pccodes = lambda lst: [d['iso_3166_1'] for d in lst]
```

```python
df['production_countries']=df['production_countries'].map(eval)
```

```python
df['pcountry_name'] = df['production_countries'].apply(get_names)
df['pcountry_isocode'] = df['production_countries'].apply(get_pccodes)
```

```python
df.drop(columns='production_countries',inplace=True)
df.head()
```

Up until now, I did all of the requested changes to the dataset.
